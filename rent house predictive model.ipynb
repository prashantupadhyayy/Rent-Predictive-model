```python

# Function to convert 'out of' values to float
def convert_out_of_to_float(floor_string):
    try:
        parts = floor_string.split(' out of ')
        if len(parts) == 2:
            return float(parts[0]) / float(parts[1])
        else:
            return None
    except ValueError:
        return None  # Handle non-numeric values like 'Upper Basement out of 4', 'Lower Basement out of 2', etc.

# Apply the function to convert 'Floor' column to float
df['Floor'] = df['Floor'].apply(convert_out_of_to_float)

# Display the DataFrame after converting 'out of' to float
print(df)
```

```python
df = df.dropna(subset=['Floor'])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
# 'Rent' is the column we want to predict

# Extract features (all columns except 'Rent' and other columns you want to exclude)
train = df.drop(['Rent', 'Posted On', 'Point of Contact', 'Tenant Preferred', 'Area Type', 'Area Locality', 'City', 'Furnishing Status'], axis=1)

# Extract target variable ('Rent')
test = df['Rent']
```

```python
from sklearn.model_selection import train_test_split

# Assuming 'train' contains your features and 'test' contains your target variable
x_train, x_test, y_train, y_test = train_test_split(train, test, test_size=0.3, random_state=2)
```

```python
regr=LinearRegression()
```

```python
regr.fit(x_train,y_train)
```

```python
pred=regr.predict(x_test)
```
